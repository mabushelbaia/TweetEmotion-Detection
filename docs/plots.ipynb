```python

import matplotlib.pyplot as plt
import numpy as np
# Example data
classifiers = ['Naive Bayes(75%)', 'Random Forest(75%)', 'Support Vector Machines (75%)',
               'Naive Bayes(5-Folds)', 'Random Forest(5-Folds)', 'Support Vector Machines (5-Folds)']
color=['orange', 'purple', 'brown', 'yellow', 'lightblue', 'red']
training_times = [1, 100.4, 2*60+49, 6,13*60+16, 10*60+49]

# Set the plot size
# plt.figure(figsize=(8,6))

# Create a bar chart
bars = plt.bar(classifiers, height=training_times,width=0.3, edgecolor='black',color=color)
plt.xticks([])
# show only ticks for values according to the height of the bars
plt.yticks(np.arange(0, 1000, 100))
plt.title('Timing for Different Classifiers')
plt.xlabel('Classifier')
plt.ylabel('Timing')
# make the legend to the right of the plot
plt.legend(bars, classifiers, loc='upper left', bbox_to_anchor=(1, 1))
handles = [plt.Rectangle((0,0),1,1, color=color[i]) for i,label in enumerate(classifiers)]
# plt.legend(handles, classifiers)
plt.xticks(rotation=90)

plt.show()

```
